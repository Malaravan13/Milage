```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_data.csv")

sns.histplot(df['mpg'], bins=20, kde=True)
plt.xlabel('MPG')
plt.ylabel('Count')
plt.title('Distribution of MPG')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='model year', y='mpg', data=df, ci=None, marker='o')
plt.title('Average MPG by Model Year')
plt.show()
```

```python
sns.boxplot(x='origin', y='mpg', data=df)
plt.title('MPG by Origin')
plt.show()
```

```python
sns.boxplot(x='cylinders', y='mpg', data=df)
plt.title('MPG by Number of Cylinders')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df[['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','Age of Car']].corr(), 
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
